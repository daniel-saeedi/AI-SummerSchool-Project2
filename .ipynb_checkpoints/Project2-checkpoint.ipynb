## Visualizing the SVM Classifier for 2 features

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn import svm

# we create 40 separable points
np.random.seed(0)
X = df[['petal length (cm)','petal width (cm)']].values.reshape(-1,2)
Y = df['target'].values.reshape(-1,1)

# figure number
fignum = 1

# fit the model
for name, penalty in (('unreg', 1), ('reg', 0.05)):

    clf = svm.SVC(kernel='linear', C=penalty)
    clf.fit(X, Y)

    # get the separating hyperplane
    w = clf.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(-5, 5)
    yy = a * xx - (clf.intercept_[0]) / w[1]

    # plot the parallels to the separating hyperplane that pass through the
    # support vectors (margin away from hyperplane in direction
    # perpendicular to hyperplane). This is sqrt(1+a^2) away vertically in
    # 2-d.
    margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
    yy_down = yy - np.sqrt(1 + a ** 2) * margin
    yy_up = yy + np.sqrt(1 + a ** 2) * margin

    # plot the line, the points, and the nearest vectors to the plane
    plt.figure(fignum, figsize=(5, 5))
    plt.clf()
    plt.plot(xx, yy, 'k-')
    plt.plot(xx, yy_down, 'k--')
    plt.plot(xx, yy_up, 'k--')

    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80,
                facecolors='none', zorder=10, edgecolors='k',
                cmap=cm.get_cmap('RdBu'))
    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=cm.get_cmap('RdBu'),
                edgecolors='k')
    plt.xlabel('petal length (cm)')
    plt.ylabel('petal width (cm)')
plt.show()

```

## Tuning Hyper Parameters

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

## predicting again after tuning hyperparameters

```python
y_pred = grid.predict(X_test)
```

##  Decision Tree and Random Forest

## 3.1 : taining a dicission tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

## 3.2 : ploting decission tree

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
tree_plot=tree.plot_tree(classifier,filled=True)
```

## plotting decission tree with max depth of 3

```python
classifier_max_depth_3 = DecisionTreeClassifier(max_depth=3)
classifier_max_depth_3.fit(X_train, y_train)
fig = plt.figure(figsize=(25,20))
tree_plot=tree.plot_tree(classifier_max_depth_3,filled=True)
```

```python
y_pred = classifier_max_depth_3.predict(X_test)
print(classification_report(y_test, y_pred, target_names=labels))
```

### What is difference between the max_depth =  3 and max_depth = 6 decision tree?
Using decision trees with bigger max_depths can sometimes make the model more accurate but sometimes it may cause  overfitting and in addition to making it less accurate on test set ,takes longer time to execute which is not ideal.

##  Tuning your hyper-parameters

```python
params = {'max_leaf_nodes': list(range(2, 100)), 'min_samples_split': [2, 3, 4],'splitter' : ['best', 'random'],'max_depth' : [3, 4 , 5 , 6 ,7 ,8, 9]}
grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42), params, verbose=1, cv=3)
grid_search_cv.fit(X_train, y_train)
```

```python
fig = plt.figure(figsize=(25,20))
tree_plot=tree.plot_tree(grid_search_cv.best_estimator_,filled=True)
```

## 3.4 : Evaluating the model

```python
y_pred = grid_search_cv.predict(X_test)
```

```python
print(classification_report(y_test, y_pred, target_names=labels))
```

## 3.5 : Training a random forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier( random_state=0)
clf.fit(X_train, y_train)
```

## 3.6 : Evaluating the model

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred, target_names=labels))
```

## What is the difference between random forest and decision tree?
A decision tree combines some decisions, whereas a random forest combines several decision trees. Thus, it is a long process, yet slow. Whereas, a decision tree is fast and operates easily on large data sets, especially the linear one. The random forest model needs rigorous training.

Our decision tree and random forest performed perfectly on test set. So we can't say which of is better for sure.
